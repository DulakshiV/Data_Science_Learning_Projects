## Data Wrangling  (Data Preprocessing/ Data Cleaning)
Goal is to convert raw data to a format suitable for data analysis

### Handle Missing Values
Missing values could be represented by various means in a dataset depending on how the dataset was created. eg: NaN , 0, blank cell etc. First we need to identify the missing values.

Our dataset has '?' marks representing missing data. These should be replaced with 'NaN', which is the default missing data representation in python for better computation. 
```python
DataFrame.replace(to_replace=None, value=None, inplace=False, limit=None, regex=False, method='pad')
```
To provide 'NaN' we import numpy library

````python
import numpy as np

#Replace '?' with NaN in the dataset
df.replace("?", np.nan, inplace = True) 
df.head()
````

#### Identify missing data: 
```python
DataFrame.isnull()
```
Return a boolean same-sized object indicating if the values are NA. NA values, such as None or numpy.NaN, gets mapped to True values. Everything else gets mapped to False values. Characters such as empty strings '' or numpy.inf are not considered NA values (unless you set pandas.options.mode.use_inf_as_na = True).

````python
missing_data = df.isnull()
missing_data.head(5)
````

#### Count missing data in each column

````python
for column in missing_data.columns.values.tolist():
    #value_count()  provide the number of occurences of each unique value in a given data series
    print (missing_data[column].value_counts())
    print("")    
````

Above output indicates 7 columns contain missing data
"normalized-losses","num-of-doors","bore","stroke","horsepower","peak-rpm","price"

#### Deal with missing data
There are various approaches to handle missing data depending on the problem.

1. Attempt to recollect the missing Data
2. Remove the data corresponding to the missing value. We could either drop only the entry corresponding to the missing datapoint(row) or drop the whole variable(column). Whole columns are dropped only when majority of the entries are empty.
  -  *dropna()* 
3. Replace missing data with approximate values. Usually replaced by mean. For categorical variables, it's replaced by the most commonly occured value (mode).We could also make a guess based on domain knowledge.
  -  *replace()*

- Replace by Mean: "normalized-losses", "stroke","bore","horsepower","peak-rpm"
- Replace by mode: "num-of-doors" Since 84% sedans is four doors.
- Drop the row:"price" Since we predict price in this task

````python
#Replace by Mean

#astype() Cast a pandas object to a specified dtype
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

avg_stroke=df['stroke'].astype('float').mean(axis=0)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)

avg_bore=df['bore'].astype('float').mean(axis=0)
df["bore"].replace(np.nan, avg_bore, inplace=True)

avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
````

````python
#Replace by mode

#find frequencies of each unique value in num-of-doors columns to find the most frequent value
print(df["num-of-doors"].value_counts())
print("")

#Or Find the most frequently occured value with idxmax()
print(df["num-of-doors"].value_counts().idxmax())

df["num-of-doors"].replace(np.nan, "four", inplace=True)
````

Since our aim is to predict automobile price, we remove all rows that doesn't have a value in the price column. As per documentation;
```python
DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```
axis
- 0, or ‘index’ : Drop rows which contain missing values.
- 1, or ‘columns’ : Drop columns which contain missing value.

how
- ‘any’ : If any NA values are present, drop that row or column.
- ‘all’ : If all values are NA, drop that row or column.

subset : array-like, optional
- Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.

inplace
- inplace=True will modify the dataset directly

````python
df.dropna(subset=["price"],axis=0,inplace=True)
````

**We should reset index, because we droped two rows**

````python
df.reset_index(drop=True, inplace=True)
df.head()
````

### Data Formatting
When importing a dataset, the data type assigned to a variable might not always be correct. Hence we have to correct those.

```python
DataFrame.dtypes #Return the dtypes in the DataFrame.

DataFrame.astype(dtype, copy=True, errors='raise', **kwargs)
```
copy : bool, default True
Return a copy when copy=True (be very careful setting copy=False as changes to values then may propagate to other pandas objects).


````python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")

df.dtypes
````

2. car fuel consumption recorded in miles per gallon unit can be converted to L/100Km metric units

Apply **data transformation** to transform mpg into L/100km

### Data Normalization
Transform variables values to a similar scale.
- Simple feature scaling (Divide by Max value of the variable)
- Min-Max
- Z-score

````python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
````

### Data Binning
Grouping values of a variable. 
eg: price of a car range from 5000 to 45500. We can group this to low-price, mid-price and high-price  
Visualize the price distribution first

```python
numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
```
Return evenly spaced numbers over a specified interval. Returns num evenly spaced samples, calculated over the interval [start, stop]. The endpoint of the interval can optionally be excluded.

````python
bins = np.linspace(df['price'].min(),df['price'].max(),4) # 4 dividers are needed for 3 bins
group_names = ["Low","Medium","High"]
````

```python
pandas.cut(x, bins, right=True, labels=None, retbins=False, precision=3, include_lowest=False, duplicates='raise')
```
Use cut when you need to segment and sort data values into bins. This function is also useful for going from a continuous variable to a categorical variable. For example, cut could convert ages to groups of age ranges. 

````python
df['price-binned']= pd.cut(df['price'],bins,labels=group_names,include_lowest=True)
df[['price','price-binned']].head(20)
````

````python
df['price-binned'].value_counts()
````

````python
#Plot the distribution of each bin
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["price-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("price")
plt.pyplot.ylabel("count")
plt.pyplot.title("price bins")
````

### Convert Categorical data to Numerical data
Most statistical models in Python are not capable of using categorical (represented as string, objects) variables as inputs. Hence these are converted to numerical values.

**One-hot encoding** : We add one feature variable per each unique value of the feature we plan to convert.
These newly added variables are also called **dummy variables/indicator variables** as those don't contain any meaning but a numerical representation of a label/feature value.

````python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
````

````python
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True)

df.head()
````
